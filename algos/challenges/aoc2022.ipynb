```python
priorities = 0
for line in input3.split('\n'):
    l = len(line)
    common = set(line[:l//2]) & set(line[l//2:])
    letter = next(iter(common))
    if 'a' <= letter <= 'z':
        priorities += ord(letter)-ord('a')+1
    if 'A' <= letter <= 'Z':
        priorities += ord(letter)-ord('A')+27
print(priorities)
```

```python
def score(letter):
    if 'a' <= letter <= 'z':
        return ord(letter)-ord('a')+1
    if 'A' <= letter <= 'Z':
        return ord(letter)-ord('A')+27

priorities = 0
lines = input3.split('\n')
for group in range(len(lines)//3):
    offset = 3*group
    common = set(lines[offset]) & set(lines[offset+1]) & set(lines[offset+2])
    priorities += score(next(iter(common)))
print(priorities)
```

```python
day7 = Puzzle(2022, 7)
```

```python
input7 = """
$ cd /
$ ls
dir a
14848514 b.txt
8504156 c.dat
dir d
$ cd a
$ ls
dir e
29116 f
2557 g
62596 h.lst
$ cd e
$ ls
584 i
$ cd ..
$ cd ..
$ cd d
$ ls
4060174 j
8033020 d.log
5626152 d.ext
7214296 k
""".strip(' ').strip('\n')
```

```python
input7 = day7.input_data
```

```python
# part 2 goes here
tree = {}
path = []
lines = input7.split('\n')
for group in mit.split_when(lines, lambda x, y: y.startswith('$ ')):
    if group[0] == '$ ls':
        node = tree
        for p in path:
            node = node[p]
        for line in group[1:]:
            size, name = line.split()
            if size == 'dir':
                node[name] = {}
            else:
                node[name] = int(size)
    else: #cd
        dir = group[0].split()[2]
        if dir == '..':
            path.pop()
        else:
            node = tree
            for p in path:
                node = node[p]
            path.append(dir)
            node[dir] = {}

def traverse(tree):
    a = b = 0
    for k,v in tree.items():
        if type(v) is dict:
            a0, b0 = traverse(v)
            a += a0
            b += b0
        else:
            a += v
    if a < 100000:
        b += a
    return a, b

output = traverse(tree)
output
```

```python
day7.answer_a = output[1]
```

```python
# part 2 goes here

def total_size(tree):
    a = 0
    for k,v in tree.items():
        if type(v) is dict:
            a += total_size(v)
        else:
            a += v
    return a
 
def traverseb(tree, a, b=float('inf')):
    b = float('inf')
    for k,v in tree.items():
        if type(v) is dict:
            a0 = total_size(v)
            b0 = traverseb(v, a)
            b = min(b, b0)
            if 30000000+a-a0 <= 70000000:
                b = min(b, a0)
    return b

outputb = traverseb(tree, total_size(tree))
outputb
```

```python
day7.answer_b = outputb
```
